```python
# No need to reexecute

# Use inflation calculator
# https://data.bls.gov/cgi-bin/cpicalc.pl?cost1=30000&year1=200001&year2=201704

import requests
from bs4 import BeautifulSoup

def fun(year):
    website=("https://data.bls.gov/cgi-bin/cpicalc.pl?cost1=1.00&year1="+str(year)+"01&year2=201701")
    #print(website)
    text = requests.get(website).text
    soup = BeautifulSoup(text,"lxml")
    tmp = soup.find_all(id="answer")
    for TMP in tmp:
        tmp2=TMP.string[1:]
        #print(TMP.string[1:])
    return tmp2

print("year,value")
for i in range(1913,2017,1):
#for i in range(1913,1916,1):
    ans=fun(i)
    print(str(i)+","+str(ans))
```

```python
# Show inflation

inflation = (pd.read_csv('inflation.csv')).set_index('year')
display(inflation)
inflation.plot()
inflationDic = inflation['value'].to_dict()
```

```python
# Create budget$2017
def fun(row):
    year = row['title_year']
    if year < 1913 or year > 2017:
        print('error, year doesnt work', year)
        print(row)
    if np.isnan(year):
        return np.nan
    
    return row['budget']*inflationDic[year]


df['budget$2017'] = df.apply(fun,axis=1)

df[['budget$2017','budget','title_year']]
```

```python
# features that need to be dropped or modified

# Can select most frequent, and then have Other category:
# genres, language, country, content_rating

# Don't know what to do yet, but might be able to construct something:
# director_name, actor_2_name,  actor_1_name, movie_title, actor_3_name, 
# plot_keywords

# Need to be dropped for sure: movie_imdb_link
df = df.drop('movie_imdb_link',axis=1)
```

```python
# No need to reexecute

# take care of : genres, language, country, content_rating

print('genres')
display(df.groupby('genres').size())
print('language')
display(df.groupby('language').size())
print('country')
display(df.groupby('country').size())
print('content_rating')
display(df.groupby('content_rating').size())
# G = TV-G        TV-Y
# PG = TV-Y7 = TV-PG = M = GP
# PG-13 = TV-14
# NC-17 = R = TV-MA = X
# Unrated = Not Rated
# Approved = Passed
```

```python
# Do genre

def fun(x, genre):
    if genre in x.split('|'):
        return 1
    else:
        return 0
    
genres = [i.split('|') for i in df.groupby('genres').size().index.values]
genres = set([item for sublist in genres for item in sublist])
print(genres)

for genre in genres:
    print(genre)
    df[genre] = df['genres'].apply(fun,args=(genre,))
    print(df[genre].sum())
    display(df[genre])
    if df[genre].sum() < 10: # Minimum kept Western at 97, the removed are at 1 to 6
        df = df.drop(genre,axis=1)
        
df = df.drop('genres',axis=1)
```

```python
# Do language

df['english']=(df['language']=='English').map({True:1,False:0})
display(df['english'])
df = df.drop('language',axis=1)
```

```python
# country
def fun(x, country):
    if x == country:
        return 1
    else:
        return 0
    
countries= df.groupby('country').size()
countries = countries[countries > 60].index.values # Biggest removed is 55
for country in countries:
    df[country] = df['country'].apply(fun,args=(country,))
    #print(country)
    #display(df[country])
    
df = df.drop('country',axis=1)
```

```python
# Content_rating
# G = TV-G        TV-Y
# PG = TV-Y7 = TV-PG = M = GP
# PG-13 = TV-14
# NC-17 = R = TV-MA = X
# Unrated = Not Rated
# Approved = Passed

df['G'] = df['content_rating'].apply(lambda x: x in ['G','TV-G','TV-Y']).map({True:1,False:0})
df['PG'] = df['content_rating'].apply(lambda x: x in ['PG','TV-Y7','TV-PG','GP','M']).map({True:1,False:0})
df['PG13'] = df['content_rating'].apply(lambda x: x in ['PG-13','TV-14']).map({True:1,False:0})
df['R'] = df['content_rating'].apply(lambda x: x in ['NC-17','R','TV-MA','X']).map({True:1,False:0})
df['U'] = df['content_rating'].apply(lambda x: x in ['Unrated','Not Rated']).map({True:1,False:0})
df['A'] = df['content_rating'].apply(lambda x: x in ['Approved','Passed']).map({True:1,False:0})
# The count is 4740
print(df['G'].sum()+df['PG'].sum()+df['PG13'].sum()+df['R'].sum()+df['U'].sum()+df['A'].sum())

df = df.drop('content_rating',axis=1)
```

```python
# 3 features are after the movie being made, one of them must be the target:
# imdb_score
# The 2 others must be dropped: movie_facebook_likes, gross
# Split into train and test data. Remove imdb_score

# Obviously some of the likes have been made after the movie, so it's possible they don't work well to predict new movies.

from sklearn.model_selection import train_test_split
train, test = train_test_split(df.drop('movie_facebook_likes',axis=1).drop('gross',axis=1), test_size=0.5,
                              random_state=1) # random_state = 1 for reproducibility

x = train.drop('imdb_score',axis=1) 
display(x)
y = train['imdb_score']
display(y)
xtest = test.drop('imdb_score',axis=1)
display(xtest)
ytest = test['imdb_score']
display(ytest)
```

```python
# XGBoost is known to be the best machine learning algorithm for this kind of task.

# Try first with parameters optimized for another tasks. I will optimized them later.

import xgboost as xgb

param={'learning_rate' : 0.1955450438296506,
     'max_depth' : 6,
     'min_child_weight' : 1.,
        'gamma' : 0.7791936249481772,
     'subsample' : 0.8200814329518703,
     'colsample_bytree' : 0.82245316066529,
     'nthread' : -1,
     'colsample_bylevel' : 0.9235200762181645,
        'reg_alpha' : 0.7308280629929902,
     'reg_lambda' : 0.4987932979468526,
           'n_estimators' : 30
          }
clf = xgb.XGBRegressor(**param)
predict = clf.fit(x,y).predict(xtest)
from sklearn.metrics import mean_squared_error     # Find other at: http://scikit-learn.org/stable/modules/classes.html#regression-metrics
print('score', mean_squared_error(ytest, predict))
```

```python
# Hyperopt is a good hyperparameter optimizer.
from hyperopt import hp, fmin, tpe, STATUS_OK, Trials
from time import time
from random import randint

evalmetrics = 'rmse' # root mean square error     The other choice for Regressor is "mae": mean absolute error
nrepeat = 20
repeat = range(nrepeat)

from sklearn.model_selection import KFold

dmatrix = xgb.DMatrix(x.values, y.values)

def objective(space):
    param={'learning_rate' : space['learning_rate'],
     'max_depth' : int(space['max_depth']),
     'min_child_weight' : space['min_child_weight'],
        'gamma' : space['gamma'],
     'subsample' : space['subsample'],
     'colsample_bytree' :space['colsample_bytree'],
     'nthread' : -1,
     'colsample_bylevel' :space['colsample_bylevel'],
        'reg_alpha' :space['reg_alpha'],
     'reg_lambda' :space['reg_lambda']  
          }
    error = 0
    print('objective function',(time() - start_time))
    for i in repeat:
        seed = randint(0, 100000)
        cvresult = xgb.cv(param, dmatrix, num_boost_round=1000, nfold=5,
                metrics=evalmetrics, early_stopping_rounds=10, shuffle = True, seed=seed)
        error += cvresult['test-rmse-mean'].tail(1).values[0]
    error /= nrepeat

    return{'loss':error, 'status': STATUS_OK }


space ={
        'max_depth': hp.quniform ('max_depth', 1, 14, 1),
        'min_child_weight': hp.quniform ('min_child_weight', 1, 10, 1),
        'subsample': hp.uniform ('subsample', 0.1, 1.),
        'learning_rate' : hp.uniform ('learning_rate', 0.01, 0.5),
       'colsample_bytree': hp.uniform ('colsample_bytree', 0.5, 1.),
        'colsample_bylevel': hp.uniform ('colsample_bylevel', 0.5, 1.),
        'gamma': hp.uniform ('gamma', 0., 1.),
    'reg_alpha': hp.uniform ('reg_alpha', 0., 1.),
    'reg_lambda' : hp.uniform ('reg_lambda', 0., 1.)
    }

trials = Trials()
start_time = time()
best = fmin(fn=objective,
            space=space,
            algo=tpe.suggest,
            max_evals=20,
            trials=trials)
print("--- %s seconds ---" % (time() - start_time))

best['max_depth'] = int(best['max_depth'])
print (best)

import numpy as np
start_time = time()
n_estimators = []
Error = []
for i in range(100):
    seed = randint(0, 100000)
    cvresult = xgb.cv(best, dmatrix, num_boost_round=1000, nfold=5,
                metrics=evalmetrics, early_stopping_rounds=50, shuffle = True, seed=seed)
    n_estimators.append(cvresult.shape[0])
    Error.append(cvresult['test-rmse-mean'].tail(1).values[0])
print("--- %s seconds ---" % (time() - start_time))
n_estimators_mean = np.mean(n_estimators)
n_estimators_std = np.std(n_estimators)
Error_mean = np.mean(Error)
Error_std = np.std(Error)
print(n_estimators_mean, n_estimators_std, Error_mean, Error_std)

best['n_estimators'] = int(round(n_estimators_mean))
clf = xgb.XGBRegressor(**best)
predict = clf.fit(x,y).predict(xtest)
from sklearn.metrics import mean_squared_error     # Find other at: http://scikit-learn.org/stable/modules/classes.html#regression-metrics
print('score', mean_squared_error(ytest, predict))
```
